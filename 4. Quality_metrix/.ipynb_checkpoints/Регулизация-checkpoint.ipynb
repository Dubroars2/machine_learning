```python
import numpy as np
import matplotlib.pyplot as plt


#будем использовать линейную регрессию, Регрессию Ridge со встроенной L2 регулизацией 
#и регрессию Lasso со встроенной L1 регулизацией
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
#Генерим случайный набор данных и задаем ф-ю
x = np.linspace(0,10,100)
y = x - 0.15*x**2+0.01*x**3 - 0.001*x**4+0.0001*x**5 + 0.4*np.random.randn(1,x.shape[0])[0]
x = x.reshape(-1,1)

plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.grid()
```

```python
#Разделим наши данные. Тест сайз занимает 85% для того, чтобы было хорошо видно. Тренировочных данных много
x_tr, x_ts, y_tr, y_ts = train_test_split(x, y, test_size=0.85, random_state=42)
```

```python
#Используем принцип полиноминальной регресси и увеличим кол-во признаков. Полином 7 степено
#Т.е. изначально у нас был 1 х и 1 y. То теперь у нас будет множество значение х-сов

p = PolynomialFeatures(7, include_bias=False)

#p.transform он трансформирует данные. Мы обучаем объект PolynomialFeatures, передаем выборку p.fit_transform(x).
#Он считает, сколько там признаков, как они будут называться и готовит себя, 
#чтобы потом другую выборку можно было трансформировать, а именно тестовую обучающую выборкуp.transform(x_tr
# и тестовую выборку p.transform(x_ts)
#Т.е трансформация - это он берет p.transform(x_tr), затем p.transform(x_ts) и эти данные превращает в соответствии с тем, 
#что он узнал p.fit_transform(x) здесь. 
#Изначально у нас был 1 признак. Мы сделали трансформацию и их стало 7 

X = p.fit_transform(x)
X_tr = p.transform(x_tr)
X_ts = p.transform(x_ts)
```

```python
X_tr.shape
```

```python
#Стандартизуем данные наши 
from sklearn.preprocessing import StandardScaler
```

```python
#Создаем объект StandardScaler и прогоняем наши данные 
#Стандартизация - все признаки занимают небольшой диапозон
sc = StandardScaler()

sc.fit(X_tr)
X_tr = sc.transform(X_tr)
X_ts = sc.transform(X_ts)
```

```python
#Мы определяемся с нашими данными, которые мы будем хранить в нашей обученной модели для того, 
#чтобы по ней построить как меняется график
#lg_r - тут будет храниться линейная регрессия Ridge
#lg_l - тут будет храниться линейная регрессия Lasso
#pred_r предсказания, которые сделала Ridge
#pred_r предсказания, которые сделала Lasso
# list_r, list_l коэффициент регулзации для одной и второй модели 
lg_r,lg_l,pred_r,pred_l = [], [], [], []
list_r=list(1.5**np.linspace(-20,20,100))
list_l=list(2**np.linspace(-20,20,100))
```

```python
#Тут мы начинаем эти модели обучать 

for i in range(len(list_r)):
  m_r = Ridge(alpha=list_r[i]).fit(X_tr, y_tr)
#Создаем модель Ridge регрессии, с коэф list_r и обучаем на наших данных (фит)
  m_l = Lasso(alpha=list_l[i]).fit(X_tr, y_tr)
#Создаем модель Lasso регрессии, с коэф list_r и обучаем на наших данных (фит)    
  lg_r.append(m_r)
#Обученные модели сохраняем в список
  pred_r.append(m_r.coef_)
#Сохраняем те коэффициенты, которая предала наша модель нашим признакам 
#Когда наша модель обучается, она подбирает коэф нашим признакам. И эти коэф мы сохраняем в список и потом сможем построить график
  lg_l.append(m_l)
  pred_l.append(m_l.coef_)
```

```python


plt.figure(figsize=[12,8])
plt.plot(x,y) #Возьмем наши даные, которые были изначально и построим график
for i, m in enumerate(lg_r): #пробегаемся по нашим моделям и для каждой модели строим те предсказания, котореы модель выучила 
  if i%4 == 0:
    plt.scatter(x_ts,m.predict(X_ts),s=2)
plt.xlim(0,10)
plt.ylim(0,5)
plt.grid()


"""
Синий график это наши первоначальные значения
остальное эти предсказания нашей модели  регрессия Ridge по точкам. Каждый цвет точки это модель. 
Каждое увеличение коэф регулизации - это новый цвет 
На первом этапе она очень хорошо предсказывает
Когда мы увеличивыем коэф регулизаци, то наша модель практически сводится к прямой линиии

"""
```

```python
plt.figure(figsize=[12,8])
plt.plot(x,y)
for i, m in enumerate(lg_l):
    plt.scatter(x_ts,m.predict(X_ts),s=2)
plt.xlim(0,10)
plt.ylim(0,5)
plt.grid()

"""
аналогично, только наша модель принимает прямую линию еще больше. 
Когда коэф регулязации подняли очень сильно, наша модешь делает предсказания в виду константы и прямая линия получается 

"""
```

```python
#Посомотрм на сами коэффициенты, что с ними происходило, когда мы повышали коэф регулязации
#Сначала были большие и потом все уменьшались и уменьшались 

plt.figure(figsize=[14,6])
x_l = np.linspace(0,len(pred_r),len(pred_r))
for i in np.vstack(pred_r).T:
    plt.plot(x_l,i)
plt.grid()    


```

```python
#Если посмотреть на маленьком промежутке, то видно, в конце, когда коэф очень большой, значения очень близки к 0 


plt.figure(figsize=[14,6])
x_l = np.linspace(0,len(pred_r),len(pred_r))
for i in np.vstack(pred_r).T:
    plt.plot(x_l,i)
plt.ylim(-0.07,0.07)    
plt.grid()
```

```python
# Аналогично для другой регресии Лессо. Они обнуляются в конце вообще полностью 

plt.figure(figsize=[14,6])
x_l = np.linspace(0,len(pred_l),len(pred_l))
for i in np.vstack(pred_l).T:
    plt.plot(x_l,np.sign(i)*np.abs(i))  
plt.grid()    
```

```python

# Синий самый важный признак, он обновился последним. А коричневый не такой важный, он первый обнулился 


plt.figure(figsize=[14,4])
x_l = np.linspace(0,len(pred_l),len(pred_l))
for i in np.vstack(pred_l).T:
    plt.plot(x_l,np.sign(i)*np.abs(i))
plt.ylim(-0.00001,0.00001)   
plt.grid() 
```

```python
# Посмотрим другой дата сет для диабетиков
# Там решается задача регресии 

from sklearn import datasets, linear_model
from sklearn.model_selection import cross_val_score
diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target
```

```python
lg_r,lg_l,pred_r,pred_l = [], [], [], []
list_r=list(1.5**np.linspace(-20,20,100))
list_l=list(2**np.linspace(-20,20,100))
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
for i in range(len(list_r)):
  m_r = Ridge(alpha=list_r[i]).fit(X, y)
  m_l = Lasso(alpha=list_l[i]).fit(X, y)
  lg_r.append(m_r)
  pred_r.append(m_r.coef_)
  lg_l.append(m_l)
  pred_l.append(m_l.coef_)
```

```python
plt.figure(figsize=[14,6])
x_l = np.linspace(0,len(pred_r),len(pred_r))
for i in np.vstack(pred_r).T:
    plt.plot(x_l,i)
plt.grid()  
```
