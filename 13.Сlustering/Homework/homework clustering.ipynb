```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel("/Users/dubrov_arseniy/Desktop/Нетология/Data Scientist/ML/13.Сlustering/Homework/geo.xlsx", 
                   index_col = 0)

df.head(2)
```

```python
# Построим график х и у

plt.scatter(df['x'], df['y'], s = 5, c = 'g')
```

```python
#Создадим объект KMean, зададим кол-во кластеров и обучим. Найдем центройды
kmeans = KMeans(n_clusters=10, random_state=0).fit(df)
centroids = kmeans.cluster_centers_
```

```python
#Посмтроим график с разбивкой на кластеры
plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=40, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=80)
plt.show()
```

```python
#Подбор количества кластеров для KMeans
#Будем на каждом шаге смотреть изменение inertia, и в тот момент, когда оно резко замедлится 
#(относительно предыдущего изменения) мы и остановимся
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(df)
    k_inertia.append(clf_kmeans.inertia_)
```

```python
diff = np.diff(k_inertia)
```

```python
plt.plot(ks[1:], diff)
```

```python
#Теперь можно найти оптимальное число кластеров 
diff_r = diff[1:] / diff[:-1]
```

```python
plt.plot(ks[1:-1], diff_r)
```

```python

k_opt = ks[np.argmin(diff_r)+1]
print(f'Оптимальное кол-во кластеров: {k_opt}')
```

```python
#Теперь сделаем аналогичное упражнение, но только с голосами ЗА
#Установим условие на датасет

yes_df = df.loc[df['comment_class'] == 1]

kmeans = KMeans(n_clusters=10, random_state=0).fit(yes_df)
centroids = kmeans.cluster_centers_

plt.scatter(yes_df['x'], yes_df['y'], c= kmeans.labels_.astype(float), s=40, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=80)
plt.show()
```

```python
#Теперь сделаем аналогичное упражнение, но только с голосами ПРОТИВ
#Установим условие на датасет

no_df = df.loc[df['comment_class'] == -1]

kmeans = KMeans(n_clusters=10, random_state=0).fit(no_df)
centroids = kmeans.cluster_centers_

plt.scatter(no_df['x'], no_df['y'], c= kmeans.labels_.astype(float), s=40, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=80)
plt.show()
```
